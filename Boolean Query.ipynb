```python
import os
import nltk
import nltk.corpus
import pandas as pd
import string
import numpy as np
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data= open(p)
        y=data.read()
        
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in y:
            if ele in punc:
                y = y.replace(ele, "")
        
        tokens=[t for t in y.split()]
        
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        sr= stopwords.words("english")
        clean_tokens=[]
        for token in tokens:
            
            wd=token.lower()
            if wd not in sr:
                clean_tokens.append(wd)
        
        doc_list.append(clean_tokens)
    return doc_list
    
```

```python
def vector(word):
    vect=[]
    for i in d_list:
        w=word.lower()
        if w in i:
            vect.append(1)
        else:
            vect.append(0)
    return vect
        
```

```python
tx=[]
for i in d_list:
    tx=tx+i
uniq_lst=[]
for t in tx:
    if t not in uniq_lst:
        uniq_lst.append(t)
```

```python
vector_dict={}
for i in uniq_lst:
    vector_dict[i] = vector(i)
```

```python
from nltk.tokenize import word_tokenize
```

```python
query = input('Enter your query:')
query = word_tokenize(query)
connecting_words = []
cnt = 1
different_words = []

for word in query:
    if word.lower() != "and" and word.lower() != "or" and word.lower() != "not":
        different_words.append(word.lower())
    else:
        connecting_words.append(word.lower())


if connecting_words[0] == "and":
    bitwise_op = np.array(vector_dict[different_words[0]]) & np.array(vector_dict[different_words[1]])
elif connecting_words[0] == "or":
    bitwise_op = np.array(vector_dict[different_words[0]]) | np.array(vector_dict[different_words[1]])

connecting_words.pop(0)
different_words.pop(0)
different_words.pop(0)

for word in connecting_words:
    if word == "and":
        bitwise_op = bitwise_op & np.array(vector_dict[different_words[0]])
        connecting_words.pop(0)
        different_words.pop(0)
    elif word == "or":
        bitwise_op = bitwise_op | np.array(vector_dict[different_words[0]])
        connecting_words.pop(0)
        different_words.pop(0)
    

print(bitwise_op)
        
```
