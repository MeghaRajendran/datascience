```python
x=df.drop(labels="1",axis=1)
y=df['1']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=101)
```

```python

from sklearn.linear_model import Perceptron

clf = Perceptron(eta0=1.0,max_iter=1000)
clf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(clf, x, y, scoring='accuracy', cv=cv, n_jobs=-1)

from numpy import mean
from numpy import std
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
classifier_pred=clf.predict(x_test)

from sklearn.metrics import accuracy_score
perc_acrcy=accuracy_score(y_test, classifier_pred)
print("accuracy= ",perc_acrcy)


```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score


recall_perc=recall_score(y_test, classifier_pred, average='macro')
print("recall= ",recall_perc)

precision_perc=precision_score(y_test, classifier_pred, average='macro')
print("precision= ",precision_perc)

```

##### fine tuning

```python
# define grid
grid = dict()
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
```

```python
from sklearn.model_selection import GridSearchCV

search = GridSearchCV(clf, grid, scoring='accuracy', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(x, y)
# summarize
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```

##### Here the  learning rate (eta0) '1.0' has the maximum mean accuracy which is already used in the model.

```python
grid1 = dict()
grid1['max_iter'] = [1, 10, 100, 1000, 10000]
```

```python
# define search
search1 = GridSearchCV(clf, grid1, scoring='accuracy', cv=cv, n_jobs=-1)
# perform the search
results = search1.fit(x, y)
# summarize
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```

##### Here the total number of training epochs (max_iter)  '100'   has the maximum mean accuracy

```python
from sklearn.linear_model import Perceptron

clf = Perceptron(max_iter=100)
clf.fit(x_train, y_train)
```

```python


# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(clf, x, y, scoring='accuracy', cv=cv, n_jobs=-1)

from numpy import mean
from numpy import std
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
classifier_pred=clf.predict(x_test)

from sklearn.metrics import accuracy_score
perc_acrcy=accuracy_score(y_test, classifier_pred)
print("accuracy= ",perc_acrcy)


```
