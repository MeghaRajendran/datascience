```python
import os
import nltk
import nltk.corpus
import pandas as pd
import string
import numpy as np
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data= open(p)
        y=data.read()
        
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in y:
            if ele in punc:
                y = y.replace(ele, "")
        
        tokens=[t for t in y.split()]
        
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        sr= stopwords.words("english")
        clean_tokens=[]
        for token in tokens:
            
            wd=token.lower()
            if wd not in sr:
                clean_tokens.append(wd)
        
        doc_list.append(clean_tokens)
    return doc_list
```
